```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def Kohler(temp, n_s, wet_diameter):
    A = calc_A(temp)
    B = calc_B(n_s)
    return((np.exp(A/wet_diameter))*np.exp(-B/((wet_diameter)**3)))

def invert_Kohler(sat_ratio, n_s, temp):
    A = calc_A(temp)
    B = calc_B(n_s)
    
    
Kohler_vec = np.vectorize(Kohler)
```

```python

MOLAR_MASS_NACL = 58.44 ##g/mol 
MOLAR_MASS_AMMONIUM = 132.14 ##g/mol

VH_NACL = 2
VH_AMMONIUM = 3

```

```python
###Define thermodynamic constants
M_w = .018 ##kg * mol^-1 molar mass of water
sigma_w = .076 ###Nm^-1
rho_w = 1000 ##kg * m^-3 density of water
R = 8.314 ###J mol^-1 K^-1 unviersal gas constant 

TEMP = 273.15 

def calc_A(temp):
    return((4*M_w*sigma_w)/(R*temp*rho_w))
def calc_B(n_s):
    return((6*n_s*M_w)/(np.pi*rho_w))

##Solute mass and molar mass expected in units of g and g/mol, respectively 
def get_ns(mass_solute, vant_hoff, molar_mass): 
    return(vant_hoff * (mass_solute/molar_mass))
```

```python
diameters = np.linspace(10**-7, 10**-5,2000)
```

```python
#List of Tuples of form (Name, ns, mass_solute)
TO_PLOT = [('NaCL', get_ns(10**-19*(10**3), VH_NACL, MOLAR_MASS_NACL), 10**-19, '-','red'), 
           ('NaCL', get_ns(10**-18*(10**3), VH_NACL, MOLAR_MASS_NACL), 10**-18, 'dotted', 'red'),
           ('Ammonium Sulfate',  get_ns(10**-19*(10**3), VH_AMMONIUM, MOLAR_MASS_AMMONIUM), 10**-19, '-', 'blue'),
           ('Ammonium Sulfate',  get_ns(10**-18*(10**3), VH_AMMONIUM, MOLAR_MASS_AMMONIUM), 10**-18, 'dotted', 'blue')]
```

```python
currfig, currax = plt.subplots(figsize=(8,6))
for triple in TO_PLOT:  
    currax.plot(diameters, Kohler_vec(TEMP, triple[1], diameters), label=f'{triple[0]}, MASS: {triple[2]} Kg',
               ls=triple[3], color=triple[4])
    
currax.legend()
currax.set_ylim([.98,1.02])
currax.set_xlim([10**-7, 10**-5])
currax.plot(diameters,[1]*len(diameters))
currax.set_ylabel('Equilibrum Saturation Ratio')
currax.set_xlabel('Wet Particle Diameter (m)')
currax.set_title('Kohler Curves for various masses of different solute')
plt.xscale('log')
plt.savefig('KohlerCurves.png', transparent=False, facecolor='w')
```
